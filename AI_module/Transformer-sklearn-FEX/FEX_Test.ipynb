```python
import matplotlib.pyplot as plt
```

```python
DATA = "/content/beckham_sample.csv"
```

```python
import pandas as pd 
df = pd.read_csv(DATA)
em_cols = ["anger","disgust","fear","happiness","sadness","surprise","neutral"]
#df = df.set_index("frame")
emo = df[em_cols + ["frame"]]
```

```python
au_cols = [x for x in df.columns if x.startswith("AU")]
aus = df[au_cols + ['frame']]
aus = aus.stack().reset_index()
aus.columns = ['frame','gesture','value']
aus1 = aus[aus['gesture'] != 'frame']

import plotly.express as px
fig = px.line_polar(aus1,
                    r='value',
                    theta='gesture',
                    line_close=True,
                    animation_frame='frame',
                    animation_group = 'gesture')
fig.show()
```

```python
tst = emo.stack().reset_index()
tst.columns = ['frame','emotion','value']
```

```python
tst1 = tst[tst['emotion'] != 'frame']
```

```python
import plotly.express as px
fig = px.line_polar(tst1,
                    r='value',
                    theta='emotion',
                    line_close=True,
                    animation_frame='frame',
                    animation_group = 'emotion')
fig.show()
```
